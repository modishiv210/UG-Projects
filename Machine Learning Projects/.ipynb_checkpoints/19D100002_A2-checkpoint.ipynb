### 2.   Write a python code to input a dataset (csv file) of Discrete datatype, automatically identify and print the

    a. Number of attributes and number of data points
    
    b. Data scale of each attribute
    
    c. Data value of each attribute

#### Note: The number of categories for categorical (nominal,ordinal) data scale for the evaluating datasets is not greater than 8.

```python
def data_description(filename):     # To get data-scale and data-value for each column in file and also the dataframe for the dataset

    
    df=pd.read_csv(filename)  # Dataframe creation
    data_value=[]
    data_scale=[]
    
  #For Data-Scale
    for col in df:
        if(len(df[col].unique())>7):
            if(min(df[col]<0)):
                data_scale.append("interval")
            else:
                data_scale.append("ratio")
        elif(len(df[col].unique())==2 or len(df[col].unique())==3):
            data_scale.append("nominal")
        else:
            data_scale.append("ordinal")
    
    #For Data-Value
    for col in df:  
        if(df.dtypes[col]=="float64" ):
             data_value.append("continuous")
        else:
             data_value.append("discrete")

    return data_scale, data_value, df
            
filename= "A2_example_dataset - students_demo.csv"

data_scale, data_value, df=data_description(filename)
print(f"This dataset contains {len(df)} entries, each with {len(df.columns)} attributes\n")
columns=df.columns
for i in range(len(data_scale)):
    print(f"{i} {columns[i]} : {data_scale[i]}, {data_value[i]}")

    
    
    
```

### 3. Write a python code to perform statistics on each data column based on its data scale.

```python
def stats(filename):   # To get all the relevant dataset for each column

    data_scale, data_value, df = data_description(filename)
    df = pd.read_csv(filename)
    entries = df[df.columns[0]].count()
    attribute_list = list(df.columns)
    
    for i in range(len(attribute_list)):
        if data_scale[i] == 'ratio':
            print(attribute_list[i])
            mode = float(df[attribute_list[i]].mode()[0])
            print("mode      {}".format(mode))
            print(df[attribute_list[i]].describe(include = 'all'), "\n")

        else:
            print(attribute_list[i])
            mode = df[attribute_list[i]].mode()[0]
            print("mode {}".format(mode))
            if data_scale[i] == 'ordinal':
                median = df[attribute_list[i]].median()
                print("median {}".format(median))

        for val in df[attribute_list[i]].unique():
            col_list = list(df[attribute_list[i]])
            print("{}      {}".format(val, col_list.count(val)))
            print("Name: {}, dtype: {} \n".format(attribute_list[i], df[attribute_list[i]].dtype))
      

    return
stats('A2_example_dataset - students_demo.csv')
```

### Write your answer here.

###   5.  Write a python code to visualize the data (a generalized method depending on data scale) 

```python
def visualize_data(filename):   # To plot distplot if data-scale is ratio , countplot if data-scale is nominal , boxplot if data-scale is ordinal

     
    df=pd.read_csv(filename)
    
    for i in range(len(columns)):
        x=df[columns[i]]
        if(len(x.unique())>7):
            plt.figure(i)
            ax=sns.distplot(x)
            ax.plot()
            plt.show()
        elif(len(x.unique())==2 or len(x.unique())==3):
            plt.figure(i)
            ax=sns.countplot(x)
            plt.tight_layout()
            ax.plot()
            plt.show()
        else:
            plt.figure(i)
            y=df['Marks']
            ax=sns.boxplot(x=x,y=y)
            plt.show()
            
name="A2_example_dataset - students_demo.csv"
visualize_data(name)
      
```
