```python
from opcua import Client
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

url = "opc.tcp://10.185.151.135:504"
client = Client(url)
client.connect()

x = []
y1 = []
y2 = []
y3 = []

def animate(i):

    time_node = client.get_node("ns=2;i=5")
    time_val = time_node.get_value()
    xvib_node = client.get_node("ns=2;i=2")
    xvib_val = xvib_node.get_value()
    yvib_node = client.get_node("ns=2;i=3")
    yvib_val = yvib_node.get_value()
    zvib_node = client.get_node("ns=2;i=4")
    zvib_val = zvib_node.get_value()

    x.append(time_val)
    y1.append(xvib_val[0])
    y2.append(yvib_val[0])
    y3.append(zvib_val[0])

    plt.cla()

    plt.plot(x, y1, label='X-axis Vibration')
    plt.plot(x, y2, label='Y-axis Vibration')
    plt.plot(x, y3, label='Z-axis Vibration')

    plt.xlabel("Time")
    plt.ylabel("Vibration")
    
    plt.legend(loc='upper left')

    plt.tight_layout()
    
    print(x,y1,y2,y3)

ani = FuncAnimation(plt.gcf(), animate, interval=100)

plt.tight_layout()
plt.show()
```
