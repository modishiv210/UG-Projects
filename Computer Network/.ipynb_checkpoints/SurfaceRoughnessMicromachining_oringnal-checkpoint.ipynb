```python
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import LineString
```

```python
def SurfaceRoughness( r,E,feed,mct ,save =False ):
    global figcount
    t0 = mct * r
    material = {0.35 : 'Ferrite' , 0.20 : 'Pearlite'}[mct]
    plt.figure( num=10 , figsize = [20,4])
    def path(x0 , y0 , r , alpha ,linelength , color = ''):
        g = x0 - r*np.sin(alpha)
        X = np.around(np.linspace( g , x0 + r , 1500),5)
        Y =  - 1 * (np.round((r**2) - (( X - x0 )**2),5))**0.5 + y0
        LX = np.linspace(g - 2*linelength, g , 1500 )
        LY = np.tan(alpha)*(LX - X[0]) + Y[0]
        x = np.concatenate((LX,X))
        y = np.concatenate((LY,Y))
        plt.plot(x,y , color)
        return x,y

    alpha = np.deg2rad(180 - E)
    x0 = 0
    y0 = 0
    O = path(x0 , y0 , r, alpha, linelength = r , color= 'r')
    M = path(x0 , y0 , r - t0, alpha , linelength=r, color = 'g')
    FinalX = np.array([])
    FinalY = np.array([])
    inter_old = 0
    for q in range(1 , 20):
        NO = path(x0 + q*feed, y0 , r, alpha, linelength=r , color = 'r')
        NM = path(x0 + q*feed, y0 , r - t0, alpha , linelength=r , color = 'g')
        first_line = LineString(np.column_stack(( NM[0] , NM[1])))
        second_line = LineString(np.column_stack(( O[0] , O[1] )))
        intersection = first_line.intersection(second_line)
        if intersection.is_empty == True :
            continue
        inter_new = np.around(intersection.x , 2)
        a = inter_old
        b = np.where(np.around(O[0],2) == inter_new )[0][0]
        c = np.where(np.around(NM[0],2) == inter_new )[0][0]
        FinalX = np.concatenate((FinalX , O[0][a:b] ))
        FinalY = np.concatenate((FinalY , O[1][a:b] ))
        inter_old = c
        O = NO
        M = NM

    plt.plot(FinalX,FinalY, 'k')
    plt.grid()
    # plt.ylim(-5 , -1)
    # plt.xlim(0,)
    if save:
        plt.xlabel('Direction of Feed -->',weight = 'bold', fontsize = 13)
        plt.ylabel('Surface Profile',weight = 'bold', fontsize = 13)
        plt.title('Surface Profile of '+material+'\n Edge Radius '+str(r)+'(um) Feed per fluke '+str(feed)+'(um) ECEA '+str(E)+' deg' ,weight = 'bold', fontsize = 14)
        # plt.show()
        plt.savefig(material+' R '+str(r)+' F '+str(feed)+' ECEA '+str(E)+'.png', bbox_inches = 'tight')

    z = np.where(np.round(FinalX,2) == 0.00)[0][-1]
    FinalX = FinalX[z:]
    FinalY = FinalY[z:]
    z1 = np.where(FinalY == max(FinalY))
    SX = FinalX[z1[0][0] : z1[0][-1] ]
    SY = FinalY[z1[0][0] : z1[0][-1] ]
    SMeanY = np.mean(SY)
    Ra = np.mean(np.abs(SY - SMeanY))
    plt.figure(num = 20 , figsize = [20 , 2])
    plt.plot(SX,SY)
    plt.plot([SX[0] , SX[-1]] , [SMeanY , SMeanY] , 'k--')
    plt.grid()
    if save:
        plt.xlabel('Direction of Feed -->',weight = 'bold', fontsize = 13)
        plt.ylabel('Surface Profile',weight = 'bold', fontsize = 13)
        plt.title('Ra Value for '+material+' is '+str(np.round(Ra,3))+'(um)\n Edge Radius '+str(r)+'(um) Feed per fluke '+str(feed)+'(um) ECEA '+str(E)+' deg' ,weight = 'bold', fontsize = 14)
        # plt.show()
        plt.savefig('Ra '+str(np.round(Ra,3))+' '+material+' R '+str(r)+' F '+str(feed)+' ECEA '+str(E)+'.png' , bbox_inches = 'tight' )
    return np.round(Ra,3)
```

```python
r = 5              ## Cutting Edge Radius
E = 10              ## End Cutting Edge angle
feed = 2            ## Feed/rev-flute
MCT_ratio = 0.35       ## Minimum Chip Thickness
SurfaceRoughness(r,E,feed,MCT_ratio , save = True)
```

```python
r = 5             ## Tool 1: 2 , Tool 2: 5 ## Cutting Edge Radius
E = 10              ## End Cutting Edge angle
MCT_ratio = 0.20    ## Perlite: 0.20 , Ferrite: 0.35   ## Minimum Chip Thickness
feeds = np.linspace(0, 0.1*40 , 41 )
material = {0.35 : 'Ferrite' , 0.20 : 'Pearlite'}[MCT_ratio]
print(feeds)
Ras = []
err= []
for p in range(len(feeds)):
    try:
        Ras += [SurfaceRoughness(r,E, feeds[p] ,MCT_ratio)]
    except:
        err+=[p]       
        pass
    # print(p)
print("Error points",err)
# print(Ras)
feeds = np.delete(feeds , err)
plt.figure(30)
plt.plot(feeds , Ras)
plt.xlabel("FEED (um)" , weight = 'bold' , fontsize = '13' )
plt.ylabel("Ra (um)", weight = 'bold' , fontsize = '13')
plt.grid()
plt.title('Ra vs Feed per fluke for '+material+'\n Edge Radius '+str(r)+'(um) ECEA '+str(E)+' deg' ,weight = 'bold', fontsize = 14)
plt.savefig('Ra vs Feed per fluke for '+material+' Edge Radius '+str(r)+'(um) ECEA '+str(E)+' deg' +'.png', bbox_inches = 'tight')
```
