```python
def maxCap(df):
    #title = 'Batt:'+arr[2]+ ' Temp:' +arr[3]+ ' DOD:' +arr[4]+ ' Dis Rate:' +arr[5]
    #columns = np.array(df.columns)  
    
    data = []
    cap = []
    discap = []
    charCycle = []
    discharCycle = []
    
    prevCap = df.loc[0][8]
    prevDisCap = df.loc[0][9]

    for i in range(len(df)):
        row = df.loc[i]
        if prevCap - row[8] < 0.1 and prevCap - row[8] > -0.1:
            charCycle.append(row[0])
            cap.append(row[8])
            prevCap = row[8]
        if prevDisCap - row[9] < 0.1 and prevDisCap - row[9] > -0.1:
            discharCycle.append(row[0])
            discap.append(row[9])       
            prevDisCap = row[9]
        
    capSOH = np.array(cap)/max(cap)
    discapSOH = np.array(discap)/max(discap)
    
    data.append(charCycle)
    data.append(capSOH)
    
    data.append(discharCycle)
    data.append(discapSOH)
    
    return data
```
