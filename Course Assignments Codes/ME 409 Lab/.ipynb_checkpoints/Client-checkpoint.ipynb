```python
import opcua
from opcua import Client
import time

import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.animation as animation

client = Client("opc.tcp://10.185.151.135:504")
client.connect()

print("Client Connected")

# Create figure for plotting
fig = plt.figure()
ax = fig.add_subplot(3, 1, 1)
ay = fig.add_subplot(3, 1, 2)
az = fig.add_subplot(3, 1, 3)

t = []
xv = []
yv = []
zv = []


# This function is called periodically from FuncAnimation
def animate(i, xs, ys):
    
    x_vib = client.get_node("ns=2;i=2")
    X_Vibration = x_vib.get_value()


    y_vib = client.get_node("ns=2;i=3")
    Y_Vibration = y_vib.get_value()


    z_vib = client.get_node("ns=2;i=4")
    Z_Vibration = z_vib.get_value()


    Time = client.get_node("ns=2;i=2")
    Time_value = Time.get_value()

    t.append(Time_value)
    xv.append(X_Vibration)
    yv.append(Y_Vibration)
    zv.append(Z_Vibration)


    t = t[-20:]
    xv = xv[-20:]
    yv = yv[-20:]
    zv = zv[-20:]


    ax.clear()
    ax.plot(t, xv)

    ay.clear()
    ay.plot(t, yv)

    az.clear()
    az.plot(t, zv)

    # Format plot
    plt.xticks(rotation=45, ha='right')
    plt.subplots_adjust(bottom=0.30)
    plt.title('Machine Vibration')
    plt.ylabel('Vibration')

# Set up plot to call animate() function periodically
ani = animation.FuncAnimation(fig, animate, fargs=(t, xv, xy, xz), interval=1000)
plt.show()
```
