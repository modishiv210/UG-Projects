```python
#Solution
 
def PrimeNumber(n):   # Function that tells whether a number is Prime or not#
  l=[]
  for i in range(2,n):
    if n%i==0:
      l.append(i)
  if len(l)==0:
    return True
  else:
    return False
 
def get_list(n):   #Function that returns the list with prime numbers less than that number
  l=[]
  for i in range(2,n):
    if PrimeNumber(i)==True:
      l.append(i)
  return l
 
n = input()
if n.isalpha() or n.isalnum() and not n.isnumeric():   #If the entered input contains alphabet then it will show this
  print("You have entered wrong input")
else :    #If the input has only numbers in it then it will call the get_list function
  a=float(n)
  b=int(a) 
  c=float(b)
  if a<0:
    print("You have entered wrong input")
  elif a<c:
    print(get_list(b+1))
  else:
    print(get_list(b))
```
